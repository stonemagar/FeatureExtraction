# make a new column for episode count



```python
df['Title'].head()
```

```python
 def extract_episodes(txt):
    check = False
    data = ""
    for i in txt:
        if i == ")":
            check = False
            return data
        if check == True:
            data = data + i
        if i == '(':
             check = True
           

```

```python
df['episodes'] = df['Title'].apply(extract_episodes)
```

```python
df.head()
```

```python
df['episodes'] = df['episodes'].str.replace("eps", "")
```

```python
df['episodes']= df['episodes'].astype(int)
```


# make a new column for time stamp



```python
def extraction_time(txt):
    check = False
    data = ""
    for i in range(len(txt)):
        if txt[i] == ')':
            for j in range(i+1,i+20):
                data += txt[j]

            return data
```

```python
df['Total_Time'] = df['Title'].apply(extraction_time)
```

```python
df.head()   
```

```python
def get_months_diff(period_str):
    start_str, end_str = period_str.split(" - ")
    start = pd.to_datetime(start_str, format="%b %Y")
    end = pd.to_datetime(end_str, format="%b %Y")
    months = (end.year - start.year) * 12 + (end.month - start.month)
    return months + 1
```

```python
df['Total_months'] = df['Total_Time'].apply(get_months_diff)
```
